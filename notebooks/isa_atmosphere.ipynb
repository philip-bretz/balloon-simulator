```python
import numpy as np
import matplotlib.pyplot as plt
from balloon_simulator.atmosphere import ISAAtmosphere

atmosphere = ISAAtmosphere()
print(atmosphere)

altitude = np.linspace(0, 10_000, 100)
res = atmosphere._calculate_state_for_altitude(altitude)
fig, axs = plt.subplots(nrows=1, ncols=3)
axs[0].plot(res.temperature, altitude)
axs[1].plot(res.pressure, altitude)
axs[2].plot(res.density, altitude)
axs[0].set_ylabel("Altitude [m]")
axs[0].set_xlabel("Temperature [K]")
axs[1].set_xlabel("Pressure [hPa]")
axs[2].set_xlabel("Density [kg/m^3]")
fig.tight_layout()
plt.show()
```
