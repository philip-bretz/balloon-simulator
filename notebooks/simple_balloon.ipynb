```python
import numpy as np
import matplotlib.pyplot as plt
from balloon_simulator.simulator import BalloonSimulator, SimulationSettings, SimulationState
from balloon_simulator.balloon import SphericalBalloon

balloon = SphericalBalloon(burst_radius=1.0, balloon_mass=1.0, payload_mass=0.25)
simulator = BalloonSimulator(balloon=balloon)

ground_state = simulator.atmosphere.calculate_state(0.0, 0.0, 0.0, 0.0)
gas_mass = balloon.calculate_gas_mass_from_volume((4/3) * np.pi * 0.9 ** 3, ground_state.pressure, ground_state.temperature)
state_0 = SimulationState(0.0, 5.43)

results = simulator.simulate(initial_time=0.0, initial_state=state_0, gas_mass=gas_mass, sim_settings=SimulationSettings(time_max=7200.0, timestep=1.0, verbose=True))

fig, ax = plt.subplots()
ax.plot(results.time, results.speed)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(results.time, results.altitude)
ax.axhline(y=0, c="gray", lw=1)
plt.show()
```
